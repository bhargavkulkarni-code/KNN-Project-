# advertising dataset doesnot have null values

```python
numerical=[fea for fea in advertising.columns if advertising[fea].dtypes!=object]
numerical
```

```python
categorical=[fea for fea in advertising.columns if advertising[fea].dtypes==object]
categorical
```

```python
advertising_numerical=advertising[numerical]
advertising_numerical
```

# advertising dataset doesnot have categorical data

```python
advertising_numerical['Sales'].max()
```

```python
x=advertising_numerical.iloc[:,1:4]
w=advertising_numerical['Sales']
bins=[0,5,10,15,20,25,30]
label=['0-5','5-10','10-15','15-20','20-25','25-30']
target=pd.cut(w,bins,labels=label)
```

```python
x
```

```python
y=target
y
```

# here x is independent and y is dependent variable here bins created for Sales.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=12)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
y_prediction=knn.predict(x_test)
y_prediction
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_prediction,digits=3))
```
